```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

#csvファイルの読み込み
df = pd.read_csv('sample_pandas_6.csv')

#CSVの読み込み
category_df = pd.read_csv('category.csv')
category_df

#マージの実施
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']],how='inner', on='商品番号')
df

#カテゴリー列の要素の出現要素の抽出
category_df=df['カテゴリー'].value_counts()
category_df

#結果を棒グラフに表示
category_df.plot(kind='bar')
plt.xlabel('カテゴリー')
plt.ylabel('出現頻度')

#注文数の基本統計量の算出
df.groupby('商品番号')['注文数'].describe()
```
